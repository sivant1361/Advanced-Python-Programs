```python
import seaborn as s
```

```python
%matplotlib inline
```

```python
t=s.load_dataset('tips')
```

```python
t
```

```python
s.distplot(t["total_bill"],kde=0.8,bins=10)
```

```python
s.jointplot(t["tip"],t["total_bill"],kind="kde")
```

```python
s.jointplot(t["tip"],t["total_bill"],kind="hex")
```

```python
s.jointplot(t["tip"],t["total_bill"],kind="reg")
```

```python
s.jointplot(t["tip"],t["total_bill"],kind="resid")
```

```python
s.jointplot(t["tip"],t["total_bill"],kind="scatter")
```

```python
s.pairplot(t,kind="reg")
```

```python
s.pairplot(t,vars=["tip","total_bill"],kind="scatter")
```
