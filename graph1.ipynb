```python
import matplotlib.pyplot as p
```

```python
#%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(1,50,10)
y=x**2
```

```python
p.plot(x,y,"r",marker="X")
p.xlabel("X-axis")
p.ylabel("Y-axis")
p.title("Title")
```

```python
p.subplot(3,2,1)
p.plot(x,y,"r")
p.subplot(3,2,(2,4))
p.plot(x,x,"y")
p.subplot(3,2,(3,5))
p.plot(y,y,"pink")
p.subplot(3,2,6)
p.plot(x**2,y**2,"g")
p.tight_layout()
```
