```python
import seaborn as s
import pandas as p
```

```python
df=p.read_csv("Salaries.csv")
df.head()
```

```python
c=df.corr()
```

```python
s.heatmap(c,linewidth=5,linecolor="white",cmap="magma",annot=True)
```
