```python
import matplotlib.pyplot as p
```

```python
%matplotlib inline
x=[2,3,5,7,11,13]
y=[2,4,16,32,64,128]
z=[]
for i in x:
    z.append(i**2)
```

```python
f=p.figure()
a1=f.add_axes([0,0,2,1.8])
a1.set_xlim(0,15)
a1.set_ylim(0,175)
a1.plot(x,y,color="blue",marker="o",markersize=13,markerfacecolor="yellow",markeredgecolor="red",markeredgewidth=3,lw=5,ls="--",alpha=0.8,label='Square')
a1.plot(x,z,color="green",marker="^",markersize=13,markerfacecolor="yellow",markeredgecolor="red",markeredgewidth=3,lw=5,ls="--",alpha=0.8,label='circle')    
a1.grid(True)
a1.legend(loc="upper left")
```
