```python
import matplotlib.pyplot as p
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[10,2,30,4,50]
```

```python
fig=p.figure()
axis1=fig.add_axes([0,0,2,2])
axis2=fig.add_axes([0.75,1.16,0.5,0.5])
axis3=fig.add_axes([1.028,1.2,0.2,0.2])
axis1.plot(x,y,lw=5,label="biggest")
axis2.plot(y,x,"r",lw=2.5,ls="--",label="bigger")
axis3.plot(x,x,"g",marker="o",markersize=7,markerfacecolor="yellow",markeredgecolor="red",label="big")
axis1.legend(loc="upper left")
axis2.legend()
axis3.legend()
```

```python
p.bar(x,y,label=['a','e','i','o','u'])
```
