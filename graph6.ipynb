```python
import seaborn as s
import pandas as pd
import numpy as n
%matplotlib inline
```

```python
w=pd.read_csv("weather1.csv")
```

```python
w.head()
```

```python
t=pd.DataFrame(w)
```

```python
s.barplot(x="Day",y="Temperature ",data=t.head(20))
```

```python
s.boxplot(x="Event",y="Temperature ",data=t)
```

```python
s.barplot(x="Event",y="Temperature ",data=t)
```

```python
s.violinplot(t["Event"],t["Temperature "],split=True)
```

```python
s.violinplot(t["Event"],t["Wind Speed"])
```

```python
s.barplot(x="Event",y="Temperature ",data=t,estimator=n.std)
```

```python
s.barplot(w["Temperature "],w["Event"])
```

```python
s.distplot(w["Temperature "])
```

```python
s.jointplot(x="Temperature ",y="Wind Speed",data=w,kind="reg")
```

```python
s.pairplot(w)
```
